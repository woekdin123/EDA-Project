# Data analysis and Visualization

# 2.	Booking Dynamics Analysis:

## a.Analyze the distribution of the number of passengers, trip types, and booking completions.

```python
#  Count plot of the number of passengers associated with each booking
ax1=sns.countplot(x='num_passengers',data = df2,palette = 'tab20',ec="black")
for bars in ax1.containers:
    ax1.bar_label(bars)
plt.xlabel("Number of Passengers")
plt.title("Number of booking with respect to number of passengers")
plt.savefig("seaborn_plot1.jpg", dpi=300)
```

#### Most of the passengers travel alone

```python
# Count trip type
ax2=sns.countplot(x="trip_type", data=df2,palette = 'Set2',edgecolor = "black")
for bars in ax2.containers:
    ax2.bar_label(bars)
plt.xlabel("Trip Type")
plt.title("Count trip type")
plt.savefig("seaborn_plot2.jpg", dpi=300)
```

### Most of the passengers booked  round trip ticket

```python
mylabels = ["Booked in other day","Booked in Holiday"]
df2['booking_complete'].value_counts().plot(kind='pie',colors=['g','r'],autopct='%.2f%%',explode = [0,0.1],labels=mylabels,shadow=True,
                                            title="Booking completions piechart")
plt.savefig("seaborn_plot3.jpg", dpi=300)
```

### Holiday booking is significantly less as compared to booking on other days 

## b.	Investigate patterns related to purchase leads, flight durations, and lengths of stay.

```python
# Correlation between purchase leads, flight durations, and lengths of stay
df3=df2[["purchase_lead","flight_duration","length_of_stay"]]
sns.heatmap(df3.corr(),annot=True,cmap="coolwarm")
plt.savefig("seaborn_plot4.jpg", dpi=300)
plt.title("Correlation between purchase leads, flight durations, and lengths of stay")
```

# Purchase lead, flight duration and duration of stay is almost independent there is a little positive correlation between flight  duration and length of stay

# 3.	Temporal Analysis:

## a.	Explore trends based on flight hours and flight days.
## b.	Investigate if there are peak times or days for bookings and travel.

```python
plt.figure(figsize = (6,4))
ax3 = sns.countplot(data=df2,x="flight_day",palette = 'deep',ec='black')
for i in ax3.containers:
    ax3.bar_label(i)
plt.savefig("seaborn_plot5.jpg", dpi=300)
```

# number of people who  travel on monday are  more  as compared to other days 
# number of people who travel on saturday are less as compare to other days

```python
plt.figure(figsize = (16,4))
ax4 = sns.countplot(data=df2,x="flight_hour",palette = 'rocket',ec='black')
for i in ax4.containers:
    ax4.bar_label(i)
plt.savefig("seaborn_plot6.jpg", dpi=300)
```

# passengers preferred to travel between 2AM to 2PM as maximum number of people travel in  that time

# 4.	Sales and Marketing Insights:

## a.	Analyse the sales channels to understand how customers reach the airline’s website.

```python
ax5=sns.countplot(data=df2,x='sales_channel',palette='Set2',ec='black')
for i in ax5.containers:
    ax5.bar_label(i)
```

# most of the people book their airline ticket through internet

```python
mylabels = ["Internet","Mobile"]
df2['sales_channel'].value_counts().plot(kind='pie',colors=['orange','green'],autopct='%.2f%%',explode = [0,0.1],labels=mylabels)
```

## b.	Investigate the geographical distribution of bookings through the booking origin.

```python
plt.figure(figsize = (15,7))
country_counts = df2.booking_origin.value_counts()
top_origin = country_counts.head(10)
#other_count = country_counts.iloc[20:].sum()

temp = pd.DataFrame({
    'Country': top_origin.index,
    'Count': top_origin.values
})
ax6=sns.barplot(x='Count', y='Country', data=temp, palette = 'rocket')
for i in ax6.containers:
    ax6.bar_label(i)
plt.title("Top 10 booking origin(Country)")
```

# most of the people take their flight from australia

# 5.	Customer Preferences:

## a.	Explore customer preferences for extra baggage, preferred seating, and in-flight meals.

```python
#pie char for visualizing percentage of each preference
plt.figure(figsize = (13,7))
plt.subplot(1,3,1)
df2['wants_extra_baggage'].value_counts().plot(kind = 'pie',autopct='%.2f%%',colors =['orange','green'])
plt.title("wants_extra_baggage")
plt.ylabel('')
plt.subplot(1,3,2)
df2['wants_preferred_seat'].value_counts().plot(kind = 'pie',autopct='%.2f%%',colors =['green','orange'])
plt.title("wants_preferred_seat")
plt.ylabel('')
plt.subplot(1,3,3)
df2['wants_in_flight_meals'].value_counts().plot(kind = 'pie',autopct='%.2f%%',colors =['green','orange'])
plt.title("wants_in_flight_meals")
plt.ylabel('')
plt.savefig("seaborn_plot501.jpg", dpi=300)
```

##  Insights
### 1) 67% of customers expressed a desire for extra baggage allowance.
### 2) 70% of customers did not prefer premium seating.
### 3) 57% of customers did not prefer a meal in-flight.


## b.	Investigate how these preferences relate to other factors such as trip type, flight duration, and number of passengers.

```python
# Correlation Heatmap of customer Preference
corr=df2[[ "flight_duration", "num_passengers",'wants_extra_baggage','wants_preferred_seat','wants_in_flight_meals']].corr()
plt.figure(figsize=(10, 5))
sns.heatmap(corr, annot=True, cmap="magma", fmt=".2f")
plt.title('Correlation Heatmap of customer Preference')
plt.savefig("seaborn_plot502.jpg", dpi=300)
```

### 1. Want_preferred_meal and want_preferred_seat has positive correlation which is 0.32. That means they are little bit dependent to each other. 
### 2. Any data in correlation table is not closer to 1 except itself that implies preferences are not so much dependent to each other.

```python
# tripwise distribution of each preferences
plt.figure(figsize = (20,5))
plt.subplot(1,3,1)
sns.countplot(data = df2,x = 'trip_type', hue = 'wants_extra_baggage')
plt.subplot(1,3,2)
sns.countplot(data = df2,x = 'trip_type', hue = 'wants_preferred_seat')
plt.subplot(1,3,3)
sns.countplot(data = df2,x = 'trip_type', hue = 'wants_in_flight_meals')
plt.savefig("seaborn_plot403.jpg", dpi=300)
```

### 1. In our data set number of percentage of round trip customer is much more higher than other two trips.
### 2. In each case maximum number of customers expressed desire for extra baggage allowance but in case of seat preferences and meal in-flight, maximum have no preference.


# 6.	Route Analysis:
### a.	Analyze the popularity and characteristics of different flight routes.
### b.	Investigate if specific routes have unique booking patterns or customer preferences.

```python
# finding top 10 most preffered routes
plt.figure(figsize = (11,7))
Flight_routes = df2.route.value_counts()
top_routes = Flight_routes.head(10)
temp = pd.DataFrame({
    'route': top_routes.index,
    'Count': top_routes.values
})
ax7 = sns.barplot(x='Count', y='route', data=temp, palette = 'rocket')
for i in ax7.containers:
    ax7.bar_label(i)
plt.title("Top 10 most preferred routes")
plt.savefig("seaborn_plot5040.jpg", dpi=300)
```

# 7. Conversion Analysis:

## a.	Explore factors that influence booking completion.
## b.	Investigate if there are discernible patterns among bookings that did not complete.


```python
# No. of passengers successfully Booked in Holiday or Other Day'
plt.figure(figsize = (10,6))
ax10 =sns.countplot(data=df2,x='num_passengers',hue='booking_complete')
plt.title('No. of passengers successfully Booked in Holiday or Other Day')
for i in ax10.containers:
    ax10.bar_label(i)
plt.savefig("seaborn_plot508.jpg", dpi=300)
```

```python
df8 = df2[df2['booking_complete'] == 0] # dataset of other day booking
df9 = df2[df2['booking_complete'] == 1] # dataset of holiday booking
```

```python
plt.figure(figsize = (13,7))
plt.subplot(1,2,1)
num_passengers_0 = df8['num_passengers'].apply(lambda x: 'other' if x>4 else x)
num_passengers_0.value_counts().plot(kind = 'pie',autopct='%.2f%%')
plt.title('Passengers Booking on others Day ')
plt.ylabel('')
plt.subplot(1,2,2)
num_passengers_1 = df9['num_passengers'].apply(lambda x: 'other' if x>4 else x)
num_passengers_1.value_counts().plot(kind = 'pie',autopct='%.2f%%')
plt.title('Passengers Booking on holiday')
plt.ylabel('')
plt.savefig("seaborn_plot509.jpg", dpi=300)
```

## In holiday number of passengers who travel alone is less as compare to number of passengers who travel alone in other day 

```python
# checking booking completion with respect to Sales Channel
ax11 = sns.countplot(data=df2,x='sales_channel',hue='booking_complete',palette =['#f21f69','#18f0da'])
plt.title('Booking completion with respect to Sales Channel')
for i in ax11.containers:
    ax11.bar_label(i)
plt.savefig("seaborn_plot510.jpg", dpi=300)
```

```python
# pie chart booking completion with respect to Sales Channel
plt.figure(figsize = (13,7))
plt.subplot(1,2,1)
df8['sales_channel'].value_counts().plot(kind = 'pie',autopct='%.2f%%',wedgeprops = {'width':0.4},pctdistance=0.4)
plt.title('Booking on Others Day ')
plt.ylabel('')
plt.subplot(1,2,2)
df9['sales_channel'].value_counts().plot(kind = 'pie',autopct='%.2f%%',wedgeprops = {'width':0.4},pctdistance=0.4)
plt.title('Booking on holiday')
plt.ylabel('')
plt.savefig("seaborn_plot511.jpg", dpi=300)
```
